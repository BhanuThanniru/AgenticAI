### Streaming and Batch

Streaming: 
    - Stream the output content while it is being generated.
    - Display the output progressively, which improves the user-experience, especially for longer responses.
    - Achieved using the stream(), which returns an iterator that yields output chunks as they are produced. Use a loop to process each chunk.

```python
#basic syntax model.stream("prompt")
model.stream("Write 150 words about AI")
```

```python
for chunk in model.stream("Write a 150 words paragraph about AI"): 
    print(chunk.text, end="", flush=True)
```

Batch:
    - collection of independent requests to a model significantly improving the performance and reduce costs as the processing can be done in parallel.

```python
responses = model.batch([
    "How are you?",
    "How is the weather today in Chicago?",
    "What are the latest trends in tech as of today?"
])

for response in responses:
    print(response.text)
```

```python
model.batch(
    ["How are you?",
    "How is the weather today in Chicago?",
    "What are the latest trends in tech as of today?"],
    config={
        "max_concurrency": 4    # Limit to 4 parallel calls
    }
)
```
