### Messages:

The input given by human is Human Message and the output given by AI is AI message.
These are fundamental unit of context for models in LangChain. They represent the input and output of models, carrying both content and metadata needed to represent the state of a conversation when interacting with an LLM.

Messages are objects that contain:
    Role: The message type (e.g: system, user)
    Content: Represents the actual content of the message (like text, images, audio, documents, etc) 
        (for AI message, basically outcome)
    Metadata: Optional fields such as response information, message IDs, and token usage

LangChain provides a standard message type that works across all model provides, ensuring consistent behavior regardless of the model being called.

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model = init_chat_model("groq:qwen/qwen3-32b")
```

```python
model.invoke("How are you doing today?") # input message given by human
```

### Text Prompts

These are strings - ideal for straightforward generation tasks where you don't need to retain conversation history.

```python
model.invoke("How are you doing today?")
```

Use text prompts when:

    You have a single, standalone request
    You don't need conversation history
    You want minimal code complexity

### Message Prompts:

Alternatively, you can pass in a list of messages to the model by providing a list of message objects.

Message types:

    System message: Instruction that tells the model how to behave and provide context for interactions
    Human message: Represents user input and interactions with the model
    AI message: Responses generated by the model, including text content, tool calls and metadata
    Tool message: Output of tool calls

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage("You are a tech expert"),
    HumanMessage("What is artificial intelligence?")
]
response = model.invoke(messages)
response.content
```

```python
# Detailed info to the LLM through System message

from langchain.messages import SystemMessage, HumanMessage

system_msg = SystemMessage("""
    You are a senior Spring Boot developer with experties in web frameworks.
    Always provide code examples and explain your reasoning.
    Be concise but thorough in explaination
""")

messages = [
    system_msg,
    HumanMessage("How to create a REST API?")
]
response = model.invoke(messages)
response.content
```

```python
# Message metadata

human_msg = HumanMessage(
    content="Hello",
    name="Alice",
    id="msg_123"
)
```

```python
response = model.invoke([
    human_msg
])
print(response)
```

```python
# AI messages

from langchain.messages import AIMessage, SystemMessage, HumanMessage

#Creating an AI message manually (e.g: conversation history)
ai_msg = AIMessage("I'd be happy to help you with that question")

# Add to conversation history
messages = [
    SystemMessage("You are a math assistant"),
    HumanMessage("Can you help me solve this?"),
    ai_msg, # insert as if it came from the model
    HumanMessage("Whats the value of 2+2?")
]
response = model.invoke(messages)
print(response.content)
```

```python
response.usage_metadata
```

```python
# Tool Message

from langchain.messages import ToolMessage, AIMessage

ai_msg = AIMessage(
    content= [],
    tool_calls = [{
        "name": "get_weather",
        "args": {"location": "Chicago"},
        "id": "call_123"
    }]
)

# Execute tool and create result message

weather_result = "Sunny, 72F"
tool_msg = ToolMessage(
    content = weather_result,
    tool_call_id = "call_123" # This should match the id in tool_calls
)

messages = [
    HumanMessage("Whats the weather in Chicago?"),
    ai_msg,
    tool_msg
]

response = model.invoke(messages)
```

```python
print(response.content)
```
