### Model integration with OpenAI, Google Gemini and Groq

```python
import os
from dotenv import load_dotenv
load_dotenv()


os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
models = init_chat_model("gpt-4.1")
models
```

```python
# call the model
response = models.invoke("Hi! How are you?")
response
```

```python
# to see the content
response.content
```

```python
### Another way to initialize model

from langchain_openai import ChatOpenAI
model = ChatOpenAI("gpt-4.1")
response = model.invoke("How are you?")
response.content
```

### Google Gemini Model Integration

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
response = model.invoke("How are you doing today?")
response
```

```python
# Another way

import os
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
response = model.invoke("How are you?")
response.content
```

### Groq model integration

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("Hi, how are you doing today?")
response.content
```

```python
# Another way

import os
from langchain_groq import ChatGroq
model= ChatGroq(model="qwen/qwen3-32b")
response = model.invoke("How are you?")
response.content
```
