### Tools

Models can request to call tools that perform tasks such as fetching data from a database, searching the web or running code. 

Tools are pairing of:
    1. A schema, including the name of the tool, a description, and/or argument definitions (often a JSON schema)
    2. A function or a coroutine to execute.

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("Hi, how are you doing today?")
response.content
```

```python
from langchain.tools import tool

@tool
def get_weather(location: str) -> str:
    """Get the weather at a location"""
    return f"The weather of {location} is Sunny"

model_with_tool = model.bind_tools([get_weather])
```

```python
response = model_with_tool.invoke("What's the weather in Chicago?")
print(response)

for tool_call in response.tool_calls:
    # View tool calls made by the model
    print(f"Tool: {tool_call['name']}")
    print(f"Args: {tool_call['args']}")
```
